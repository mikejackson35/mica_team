```python
#### DEPENDENCIES ####

import pandas as pd
import csv
```

```python
#### READ IN 20 YEARS OF PASSING STATS ####

file_2000 = "Resources/2000_passing.csv"
file_2000_df = pd.read_csv(file_2000)

file_2001 = "Resources/2001_passing.csv"
file_2001_df = pd.read_csv(file_2001)

file_2002 = "Resources/2002_passing.csv"
file_2002_df = pd.read_csv(file_2002)

file_2003 = "Resources/2003_passing.csv"
file_2003_df = pd.read_csv(file_2003)

file_2004 = "Resources/2004_passing.csv"
file_2004_df = pd.read_csv(file_2004)

file_2005 = "Resources/2005_passing.csv"
file_2005_df = pd.read_csv(file_2005)

file_2006 = "Resources/2006_passing.csv"
file_2006_df = pd.read_csv(file_2006)

file_2007 = "Resources/2007_passing.csv"
file_2007_df = pd.read_csv(file_2001)

file_2008 = "Resources/2008_passing.csv"
file_2008_df = pd.read_csv(file_2008)

file_2009 = "Resources/2009_passing.csv"
file_2009_df = pd.read_csv(file_2009)

file_2010 = "Resources/2010_passing.csv"
file_2010_df = pd.read_csv(file_2010)

file_2011 = "Resources/2011_passing.csv"
file_2011_df = pd.read_csv(file_2011)

file_2012 = "Resources/2012_passing.csv"
file_2012_df = pd.read_csv(file_2012)

file_2013 = "Resources/2013_passing.csv"
file_2013_df = pd.read_csv(file_2013)

file_2014 = "Resources/2014_passing.csv"
file_2014_df = pd.read_csv(file_2014)

file_2015 = "Resources/2015_passing.csv"
file_2015_df = pd.read_csv(file_2015)

file_2016 = "Resources/2016_passing.csv"
file_2016_df = pd.read_csv(file_2016)

file_2017 = "Resources/2017_passing.csv"
file_2017_df = pd.read_csv(file_2017)

file_2018 = "Resources/2018_passing.csv"
file_2018_df = pd.read_csv(file_2018)

file_2019 = "Resources/2019_passing.csv"
file_2019_df = pd.read_csv(file_2019)

file_2020 = "Resources/2020_passing.csv"
file_2020_df = pd.read_csv(file_2020)              
```

```python
#### CONCATINATE ALL 20 ####

stats_concat = pd.concat([file_2000_df, file_2001_df, file_2002_df,file_2003_df, file_2004_df, file_2005_df, file_2006_df, file_2007_df, file_2008_df, file_2009_df, file_2010_df, file_2011_df, file_2012_df, file_2013_df, file_2014_df, file_2015_df, file_2016_df, file_2017_df,file_2018_df, file_2019_df, file_2020_df])
```

```python
#### TAKE A LOOK ####

stats_concat.head(2)
```

```python
#### DROP UNWANTED COLUMNS ####

stats_concat = stats_concat.drop(['Rk', 'Pos', 'Cmp', 'Att', 'Cmp%', 'Yds', 'TD%', 'Int', 'Int%', '1D', 'Lng', 'AY/A', 'Y/C', 'Sk', 'Yds.1', 'Sk%', 'NY/A', 'ANY/A', '4QC', 'GWD', 'QBR'], axis = 1)
```

```python
stats_concat['Player'] = stats_concat['Player'].str.replace("*", '')
```

```python
#### TAKE A LOOK ####

stats_concat.head(2)
```

```python
#### READ IN STATIC_DATA CSV IN ORDER TO GET LIST OF OUR POPULATION ####

qb_population = "Resources/static_data.csv"
qb_population_df = pd.read_csv(qb_population)
list_of_88 = qb_population_df["Player"].tolist()
```

```python
#### LEAVE ONLY PLAYERS IN OUR LIST OF 88 QUARTERBACKS ####

stats_concat = stats_concat[stats_concat.Player.isin(list_of_88)].sort_values('Player')
```

```python
#### TAKE A LOOK ####

stats_concat.head(2)
```

```python
#### SEND TO CSV FOR FINAL CLEANUP (PLAYER_SEASON, SALARY, AND WIN/LOSS/TIE FORMATTING)

stats_concat.to_csv("../Resources/stats_concat_cleaned2.csv", index=False, header=True)
```
