```python
#dependencies

import pandas as pd
import csv
```

```python
#read in statistics files

file_2000 = "../mica_resources/2000_passing.csv"
file_2000_df = pd.read_csv(file_2000)

file_2001 = "../mica_resources/2001_passing.csv"
file_2001_df = pd.read_csv(file_2001)

file_2002 = "../mica_resources/2002_passing.csv"
file_2002_df = pd.read_csv(file_2002)

file_2003 = "../mica_resources/2003_passing.csv"
file_2003_df = pd.read_csv(file_2003)

file_2004 = "../mica_resources/2004_passing.csv"
file_2004_df = pd.read_csv(file_2004)

file_2005 = "../mica_resources/2005_passing.csv"
file_2005_df = pd.read_csv(file_2005)

file_2006 = "../mica_resources/2006_passing.csv"
file_2006_df = pd.read_csv(file_2006)

file_2007 = "../mica_resources/2007_passing.csv"
file_2007_df = pd.read_csv(file_2001)

file_2008 = "../mica_resources/2008_passing.csv"
file_2008_df = pd.read_csv(file_2008)

file_2009 = "../mica_resources/2009_passing.csv"
file_2009_df = pd.read_csv(file_2009)

file_2010 = "../mica_resources/2010_passing.csv"
file_2010_df = pd.read_csv(file_2010)

file_2011 = "../mica_resources/2011_passing.csv"
file_2011_df = pd.read_csv(file_2011)

file_2012 = "../mica_resources/2012_passing.csv"
file_2012_df = pd.read_csv(file_2012)

file_2013 = "../mica_resources/2013_passing.csv"
file_2013_df = pd.read_csv(file_2013)

file_2014 = "../mica_resources/2014_passing.csv"
file_2014_df = pd.read_csv(file_2014)

file_2015 = "../mica_resources/2015_passing.csv"
file_2015_df = pd.read_csv(file_2015)

file_2016 = "../mica_resources/2016_passing.csv"
file_2016_df = pd.read_csv(file_2016)

file_2017 = "../mica_resources/2017_passing.csv"
file_2017_df = pd.read_csv(file_2017)

file_2018 = "../mica_resources/2018_passing.csv"
file_2018_df = pd.read_csv(file_2018)

file_2019 = "../mica_resources/2019_passing.csv"
file_2019_df = pd.read_csv(file_2019)

file_2020 = "../mica_resources/2020_passing.csv"
file_2020_df = pd.read_csv(file_2020)
```

```python
#contcatonate all stats files into one

full_statistics_df = pd.concat([file_2000_df, file_2001_df, file_2002_df,file_2003_df, file_2004_df, file_2005_df, file_2006_df, file_2007_df, file_2008_df, file_2009_df, file_2010_df, file_2011_df, file_2012_df, file_2013_df, file_2014_df, file_2015_df, file_2016_df, file_2017_df,file_2018_df, file_2019_df, file_2020_df])
```

```python
##read to csv for changes easier to make in excel

full_statistics_df.to_csv("Output/full_statistics.csv", index=False, header=True)

##bring back from Excel

updated_full_statistics = "Output/full_stats.csv"
updated_full_statistics_df = pd.read_csv(updated_full_statistics)

##read in qb_population csv and grab list of 100 qb's

qb_population = "drafts_qb_only_all_years.csv"
qb_population_df = pd.read_csv(qb_population)

qb_100_list = qb_population_df["Player"].tolist()

##remove qb's not in our qb_100_list and set index to Player

qb_statistics_df = updated_full_statistics_df[updated_full_statistics_df.Player.isin(qb_100_list)].set_index('Player')
```
