```python
#### DEPENDENCIES ####

import pandas as pd
import csv
```

```python
#### READ IN CSV DOWNLOADED FROM SPORTS-REFERENCE.COM ####

new_qb = "Resources/new_qb.csv"
new_qb_df = pd.read_csv(new_qb)
```

```python
#### TAKE A LOOK ####

new_qb_df
```

```python
#### REMOVE PLAYERS WITH LESS THAN 9 GAME ('G') APPEARANCES IN CAREER ####

new_qb_df.drop(new_qb_df[new_qb_df['G'] < 9].index, inplace = True)
new_qb_df
```

```python
new_qb_df=new_qb_df.fillna(0)
```

```python
new_qb_df=new_qb_df.loc[new_qb_df['G'] != 0,:]
```

```python
#### DROP UNEEDED COLUMNS ###

new_qb_df = new_qb_df.drop(['Year', 'Pos', 'From', 'To', 'AP1', 'PB', 'St', 'wAV', 'G', 'GS', 'Unnamed: 16'], axis = 1)
```

```python
new_qb_df
```

```python
new_qb_df.to_csv("Resources/qb_rough.csv", index=False, header=True)
```
